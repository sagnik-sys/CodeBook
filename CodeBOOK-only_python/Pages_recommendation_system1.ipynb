# PAGES RECOMMENDATION SYSTEM
## BUT FILTERED SUCH THAT ALWAYS SHARED SCORE>0 
### I.E FOR GETTING RECOMENDATION ATLEAST THERE SHOULD BE 1 MUTUAL LIKING OF PAGE

```python
import json

#to load the data
def load_data(filename):
    with open ("massive_data.json","r") as f:
        return json.load(f)
```

```python
def find_pages_you_might_like(user_id, data):
    user_pages = {}
    
    # Populate user_pages dictionary
    for user in data['users']:
        user_pages[user['id']] = set(user['liked_pages'])

    if user_id not in user_pages:
        return []
        
    user_liked_pages = user_pages[user_id]
    page_suggestion = {}

    for other_user, pages in user_pages.items():
        if other_user != user_id:
            shared_pages = user_liked_pages.intersection(pages)
            shared_count = len(shared_pages)
            
            # Only suggest pages if there are shared interests
            if shared_count > 0:
                for page in pages:
                    if page not in user_liked_pages:
                        # Correctly use the page variable as key
                        page_suggestion[page] = page_suggestion.get(page, 0) + shared_count
     
    # Filter out pages with score <= 0 and sort
    sorted_pages = sorted(
        [(page, score) for page, score in page_suggestion.items() if score > 0],
        key=lambda x: x[1], 
        reverse=True
    )
    
    return [page_id for page_id, score_count in sorted_pages]
```

```python
data=load_data("massive_data.json")
user_id=1
page_recommendation=find_pages_you_might_like(user_id, data)
print(page_recommendation)
```
